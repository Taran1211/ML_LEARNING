# Useful for making plots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

##            Plotting the data 

plt.plot(x,y)
plt.show()

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave 
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'*')
plt.show()
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
```

# Bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Greek']
people=['100','200','50','400']
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of people')
plt.show()
```

```python
#pie chart
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Greek']
people=['100','200','50','400']
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([1,1,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
fig3=plt.figure();
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
